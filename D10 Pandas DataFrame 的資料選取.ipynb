```python
df = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
df
```

```python
print(df.loc[0:2, 'col1':'col3'])
print(df.loc[[1, 3, 5], 'col1':'col3'])
print(df.loc[0:5, 'col2':'col3'])
print(df.loc[[0, 2, 4],'col2'])
```

```python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
```

```python
print(df[df[0] > 20])
print(df[df[0]+df[1] > 50])
print(df[(df[0] < 30) | (df[1] > 30)])
print(df[df[0]+df[1]+df[2]+df[3] > 100])
```
